```python
import pandas as pd
from IPython.display import display

file_path = '/content/drive/My Drive/amazon_database_Akram.xlsx'

try:
    df = pd.read_excel(file_path, engine='openpyxl')


    display(df.head(10))

except Exception as e:
    print(f"Ошибка при загрузке файла с движком openpyxl: {e}")
```

```python
top_10_expensive = df.sort_values(by='Цена', ascending=False).head(10)
display(top_10_expensive[['Название товара', 'Цена']])
```

```python
top_10_low_prise = df.sort_values(by='Цена', ascending=True).head(10)
display(top_10_expensive[['Название товара', 'Цена']])
```

```python
top_10_sales = df.sort_values(by='Продажи за месяц', ascending=False).head(10)
display(top_10_sales[['Название товара', 'Цена', 'Продажи за месяц']])
```

```python
top_10_worst_sales = df.sort_values(by='Продажи за месяц', ascending=True).head(10)
display(top_10_sales[['Название товара', 'Цена', 'Продажи за месяц']])
```

```python
price_counts = df.groupby('Цена')['Цена'].count().reset_index(name='Количество')
popular_prices = price_counts.sort_values(by='Количество', ascending=False)
print(popular_prices.head(10))
```

```python
price_counts = df.groupby('Цена')['Цена'].count().reset_index(name='Количество')
popular_prices = price_counts.sort_values(by='Количество', ascending=True)
print(popular_prices.head(10))
```

```python
df['Popularity_Price_Ratio'] = df['Продажи за месяц'] / df['Цена']
df_sorted = df.sort_values(by=['Popularity_Price_Ratio'], ascending=False)
best_product = df_sorted.iloc[0]
print(best_product[['Название товара', 'Цена', 'Продажи за месяц']])
```
