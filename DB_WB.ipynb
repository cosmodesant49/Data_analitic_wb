```python
import pandas as pd
from IPython.display import display

file_path = '/content/drive/My Drive/amazon database Akram.xlsx'

try:
    df = pd.read_excel(file_path, engine='openpyxl')


    display(df.head(10))

except Exception as e:
    print(f"Ошибка при загрузке файла с движком openpyxl: {e}")
```

```python
top_10_expensive = df.sort_values(by='Цена', ascending=False).head(10)
display(top_10_expensive[['Название товара', 'Цена']])
```

```python
top_10_low_prise = df.sort_values(by='Цена', ascending=True).head(10)
display(top_10_expensive[['Название товара', 'Цена']])
```

```python
top_10_sales = df.sort_values(by='Продажи за месяц', ascending=False).head(10)
display(top_10_sales[['Название товара', 'Цена', 'Продажи за месяц']])
```

```python
top_10_worst_sales = df.sort_values(by='Продажи за месяц', ascending=True).head(10)
display(top_10_sales[['Название товара', 'Цена', 'Продажи за месяц']])
```

```python
price_counts = df.groupby('Цена')['Цена'].count().reset_index(name='Количество')
popular_prices = price_counts.sort_values(by='Количество', ascending=False)
print(popular_prices.head(10))
```

```python
price_counts = df.groupby('Цена')['Цена'].count().reset_index(name='Количество')
popular_prices = price_counts.sort_values(by='Количество', ascending=True)
print(popular_prices.head(10))
```

```python
df['Popularity_Price_Ratio'] = df['Продажи за месяц'] / df['Цена']
df_sorted = df.sort_values(by=['Popularity_Price_Ratio'], ascending=False)
best_product = df_sorted.iloc[0]
print(best_product[['Название товара', 'Цена', 'Продажи за месяц']])
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Цена')

plt.plot(df_sorted['Цена'], df_sorted.index)
plt.xlabel('Цена')
plt.ylabel('Индекс')
plt.title('Линейный график от самой низкой до самой высокой цены')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

rating_counts = df.groupby('Рейтинг ')['Рейтинг '].count()

plt.bar(rating_counts.index, rating_counts.values)
plt.xlabel('Рейтинг')
plt.ylabel('Количество')
plt.title('Диаграмма рейтинга')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Цена'], df['Продажи за месяц'])
plt.xlabel('Цена')
plt.ylabel('Продажи за месяц')
plt.title('Точечный график отношения цены к продажам')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Convert 'Цена' and 'Продажи за месяц' columns to string type
plt.scatter(df['Цена'].astype(str), df['Продажи за месяц'].astype(str))

plt.xlabel('Цена')
plt.ylabel('Продажи за месяц')
plt.title('Точечный график количества продаж')
plt.grid(True)
plt.show()
```
