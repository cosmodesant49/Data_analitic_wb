```python
#1
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
import numpy as np


file_path = '/content/drive/My Drive/amazon data 3.xlsx'

df = pd.read_excel(file_path, engine='openpyxl')

x = df[['цена','продажи за месяц','общее количество комментариев','рейтинг']]
sns.heatmap(x.corr(), xticklabels=x.corr().columns, yticklabels=x.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

```python
#2
sns.scatterplot(x=df['общее количество комментариев'],y=df['рейтинг'])
```

```python
#3
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='цена', y='продажи за месяц', palette='viridis')
plt.title("Цена vs Продажи ")
```

```python
#4
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='тип товара', y='рейтинг')
plt.xticks(rotation=45)
plt.title("Средний рейтинг по типу товара")
```

```python
#5
plt.figure(figsize=(8, 8))
top_8_types = df['тип товара'].value_counts().nlargest(8)
top_8_types.plot.pie(autopct='%1.1f%%')
plt.title("Распределение товаров по типу")
plt.ylabel('')
```

```python
#6
sales_by_type = df.groupby('тип товара')['продажи за месяц'].sum().sort_values(ascending=False)
top_5_sales = sales_by_type.head(5).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_sales, x='тип товара', y='продажи за месяц')
plt.title("Топ-5 типов товаров по продажам за месяц")
plt.xticks(rotation=30)
```

```python
sns.set(style="whitegrid")
plt.rcParams.update({'figure.max_open_warning': 0})

fig, axs = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle("📊 Товарный Dashboard", fontsize=20)

# 1
x = df[['цена','продажи за месяц','общее количество комментариев','рейтинг']]
sns.heatmap(x.corr(), xticklabels=x.corr().columns, yticklabels=x.corr().columns, cmap='RdYlGn', center=0, annot=True,ax=axs[0, 0])
axs[0, 0].set_title("Кореляционная карта")

# 2
sns.scatterplot(x=df['общее количество комментариев'],y=df['рейтинг'],ax=axs[0, 1])
axs[0, 1].set_title("Влияние количества комментариев на рейтинг")

# 3
sns.scatterplot(data=df, x='цена', y='продажи за месяц',ax=axs[1, 0])
axs[1, 0].set_title("влияние цен на количество продаж ")

# 4
sns.barplot(data=df, x='тип товара', y='рейтинг',ax=axs[1, 1])
plt.xticks(rotation=45)
axs[1, 1].set_title("какие рейтинги у товаров с этими типами")
axs[1, 1].tick_params(axis='x', rotation=55)

# 5
top_8_types = df['тип товара'].value_counts().nlargest(8)
top_8_types.plot.pie(autopct='%1.1f%%',ax=axs[2, 0])
plt.ylabel('')
axs[2, 0].set_title("каких типов товары в продаже")

# 6
sales_by_type = df.groupby('тип товара')['продажи за месяц'].sum().sort_values(ascending=False)
top_5_sales = sales_by_type.head(5).reset_index()
sns.barplot(data=top_5_sales, x='тип товара', y='продажи за месяц',ax=axs[2, 1])
axs[2, 1].set_title("Топ-5 типов по продажам")
axs[2, 1].tick_params(axis='x', rotation=30)



plt.tight_layout(rect=[0, 0, 1, 1])
```

```python
top_prices = df['Средняя цена по типу товара'].sort_values(ascending = False).head(5)
labels = top_prices.index
prices = top_prices.values
x_pos = np.arange(len(labels))

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(projection = '3d')

y_pos = np.zeros_like(x_pos)
z_pos = np.zeros_like(x_pos)
dx = np.ones_like(x_pos)
dy = np.ones_like(x_pos)
dz = prices

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color='cyan')

ax.set_xticks(x_pos)
ax.set_xticklabels(labels, rotation=45)
ax.set_title("Средние цены товаров по 5 типам")
ax.set_zlabel("Цена")
```
